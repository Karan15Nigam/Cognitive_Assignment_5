```python
#QUESTION 1
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
#(i)sum of all elements
total_sum = np.sum(gfg)
print(total_sum)
#(ii)sum of all elements row wise
row_sum = np.sum(gfg, axis=1)
print(row_sum)
#(iii)sum of all elements column wise
col_sum = np.sum(gfg, axis=0)
print(col_sum)
```

```python
#QUESTION 2
#(a)
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])
#(i)sorted array
sorted_array = np.sort(array)
print(sorted_array)
#(ii)indicies of sorted array
sorted_indices = np.argsort(array)
print(sorted_indices)
#(iii)4 smallest elements
smallest_4 = np.sort(array)[:4]
print(smallest_4)
#(iv)5 largest elements
largest_5 = np.sort(array)[-5:]
print(largest_5)

#(b)
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
#(i)integer elements only
integer_elements = array[array == np.floor(array)]
print(integer_elements)
#(ii)float elements only
float_elements = array[array != np.floor(array)]
print(float_elements)
```

```python
#QUESTION 3
import numpy as np

#(a): Generate sales dataset
first_initial = 'A'
last_initial = 'B'

X = ord(first_initial) + ord(last_initial)
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales array:", sales)

#(b): Compute tax rate and apply
tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 + tax_rate)
print("Sales after tax:", sales_after_tax)

#(c): Apply discount
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Discounted sales:", discounted_sales)

#(d): Expand sales data for multiple weeks
weekly_sales = np.tile(discounted_sales, (3, 1))
week_factors = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
adjusted_weekly_sales = weekly_sales * week_factors

print("Weekly sales after 2% increase per week:\n", adjusted_weekly_sales)
```

```python
#QUESTION 4
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -10 to 10 with 100 points
x = np.linspace(-10, 10, 100)

# Compute y values for each function
y1 = x ** 2                   # Y = x^2
y2 = np.sin(x)                # Y = sin(x)
y3 = np.exp(x)                # Y = e^x
y4 = np.log(np.abs(x) + 1)    # Y = log(|x| + 1)

# Create a figure with subplots
plt.figure(figsize=(8, 5))

# Plot Y = x^2
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='r', label=r'$y = x^2$')
plt.title("Quadratic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Plot Y = sin(x)
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='b', label=r'$y = \sin(x)$')
plt.title("Sine Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Plot Y = e^x
plt.subplot(2, 2, 3)
plt.plot(x, y3, color='g', label=r'$y = e^x$')
plt.title("Exponential Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Plot Y = log(|x| + 1)
plt.subplot(2, 2, 4)
plt.plot(x, y4, color='purple', label=r'$y = \log(|x| + 1)$')
plt.title("Logarithmic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
